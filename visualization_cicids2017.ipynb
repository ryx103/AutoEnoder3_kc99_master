```python
import pandas as pd

data_path = './data/MachineLearningCVE/ProcessedDataset/test_05.csv'
df = pd.read_csv(data_path)
print(df.columns)
```

```python
df.head(100)
```

```python
print(f"The combined dataset contains {df.shape[0]} instances and {df.shape[1]} features.")
```

```python
# Check for missing values
missing_values = df.isnull().sum().sum()
print(f"The dataset contains {missing_values} missing values.")

# Remove instances with missing class labels
df = df[~df['Label'].isnull()]

# Remove instances with missing information
df = df.dropna()
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"The dataset contains {duplicates} duplicate rows.")
# Check the final shape of the DataFrame
print(f"After cleaning, the dataset contains {df.shape[0]} instances and {df.shape[1]} features.")
# Count the occurrences of each class label
class_counts = df['Label'].value_counts()
print("Class label counts:")
print(class_counts)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)  
print(df.duplicated().sum(), "fully duplicate rows to remove")
df.drop_duplicates(inplace=True)
print(df.duplicated().sum(), "fully duplicate rows to remove")
```

```python
df.info()
df['Label'].value_counts()
y = df['Label']
class_dist = y.value_counts();
print(class_dist);
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
plt.bar(class_dist.index, class_dist.values)
plt.xlabel('Class')
plt.ylabel('Number of samples')
plt.title('Class distribution of CICIDS2017 dataset', fontsize=10)
plt.savefig("Class_distribution_samples.png", dpi=500)
plt.show()
```

```python
class_counts = df['Label'].value_counts()
```

```python
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(20, 10))

# Generate the barplot
sns.barplot(x=class_counts.index, y=class_counts.values, linewidth=2)

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=90, fontsize=20)

# Set labels and title with increased font size
plt.xlabel('Attack Types', fontsize=25)
plt.ylabel('Frequency', fontsize=25)
plt.title('Class Distribution in CICIDS2017 Dataset', fontsize=10)

# Adding grid
plt.grid(True)

# Save the plot as a high-resolution image
plt.savefig("Class_distribution_CICIDS2017.png", dpi=500)

# Show the plot
plt.show()
```

```python
majority_class_count = class_counts.max();
minority_class_count = class_counts.min();
imbalance_ratio = majority_class_count / minority_class_count;
print(f"Imbalance Ratio: {imbalance_ratio:.2f}");
```

```python
# Count the occurrences of each class label
class_counts = df['Label'].value_counts()

# Calculate the percentages
total = class_counts.sum()
percent_total = class_counts / total * 100
percent_majority = class_counts / class_counts['BENIGN'] * 100

# Create a DataFrame for the table
table = pd.DataFrame({
    'Normal / Attack Labels': class_counts.index,
    'Number of Instances': class_counts.values,
    '% of Prevalence w.r.t. the Majority Class': percent_majority.values,
    '% of Prevalence w.r.t. the Total Instances': percent_total.values
})

table.reset_index(inplace=True)
table.rename(columns={'index': 'Sl No'}, inplace=True)
print(table)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a bar plot for the number of instances per class
plt.figure(figsize=(20, 10))
sns.barplot(x='Normal / Attack Labels', y='Number of Instances', data=table)
plt.xticks(rotation=90)
plt.title('Number of Instances per Class in CICIDS2017 DATASET', fontsize=10)
plt.savefig("Number_of_Instances_per_Class_CICIDS2017.png", dpi=500)
plt.show()
```

```python
# Create a pie chart for the prevalence w.r.t. the total instances
plt.figure(figsize=(20, 20))
plt.pie(table['% of Prevalence w.r.t. the Total Instances'], labels=table['Normal / Attack Labels'], autopct='%1.1f%%')
plt.title('% Prevalence w.r.t. the Total Instances in CICIDS2017 Dataset', fontsize=10)
plt.savefig("Prevalence_wrt_the_Total_Instances.png", dpi=500)
plt.show()
```

```python
# Compute the correlation matrix
corr = df[['Flow Duration', 'Flow Bytes/s', 'Fwd Packet Length Mean', 'Bwd Packet Length Mean']].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 8))

# Generate a custom diverging colormap
cmap = "coolwarm"

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmin=-1, vmax=1, center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": .5}, annot=True, annot_kws={"size": 12})

# Increase font size
plt.title('Correlation Heatmap of Selected Features in CICIDS2017 Dataset', fontsize=10)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Save as high-resolution image
plt.savefig("Corr_Heatmap_SelectedFeatures.png", dpi=500)

plt.show()
```

```python
df['Label'] = pd.to_numeric(df['Label'], errors='coerce')
# Correlation Matrix Heatmap

corr_matrix = df.corr()

# Increase the size of the figure
plt.figure(figsize=(50, 50))

# Generate the heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', linewidths=0.5, cmap='coolwarm', annot_kws={"size": 10}, cbar_kws={"shrink": .5})

# Increase font size of title and axis labels
plt.title('Correlation Matrix Heatmap of CICIDS2017 Dataset', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Save the figure in high resolution
plt.savefig("Corr_Matrix_Heatmap_Dataset.png", dpi=500)

# Show the plot
plt.show()
```
